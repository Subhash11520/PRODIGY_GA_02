```python
# ========== 📌 Step 1: Install Required Libraries ==========
!pip install -q transformers diffusers accelerate
!pip install -q torch torchvision

# ========== 📌 Step 2: Mount Google Drive ==========
from google.colab import drive
drive.mount('/content/drive')

# ========== 📌 Step 3: Import Libraries ==========
import torch
from diffusers import StableDiffusionPipeline
from PIL import Image
import matplotlib.pyplot as plt

# ========== 📌 Step 4: Load Pretrained Stable Diffusion Model ==========
model_id = "runwayml/stable-diffusion-v1-5"

pipe = StableDiffusionPipeline.from_pretrained(
    model_id,
    torch_dtype=torch.float16,
    revision="fp16"
).to("cuda")

# ========== 📌 Step 5: Prompt for Image Generation ==========
prompt = "a futuristic city floating in the clouds, ultra detailed, 4K"

image = pipe(prompt).images[0]

# ========== 📌 Step 6: Show the Generated Image ==========
plt.imshow(image)
plt.axis('off')
plt.title(prompt)
plt.show()

# ========== 📌 Step 7: Save to Google Drive ==========
output_path = "/content/drive/MyDrive/Generated_Image.png"
image.save(output_path)
print(f"✅ Image saved to: {output_path}")
```

```python
# 📌 Step 1: Install required packages
!pip install -q diffusers transformers accelerate safetensors
!pip install -q torch torchvision

# 📌 Step 2: Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# 📌 Step 3: Import libraries
import torch
from diffusers import StableDiffusionPipeline
import matplotlib.pyplot as plt

# 📌 Step 4: Load model (no fp16 revision)
model_id = "runwayml/stable-diffusion-v1-5"

pipe = StableDiffusionPipeline.from_pretrained(
    model_id,
    torch_dtype=torch.float16,
    use_safetensors=True
).to("cuda")

# 📌 Step 5: Prompt for generation
prompt = "a fantasy landscape with floating islands, 4K, digital art"

# Generate image
image = pipe(prompt).images[0]

# 📌 Step 6: Display the result
plt.imshow(image)
plt.axis('off')
plt.title(prompt)
plt.show()

# 📌 Step 7: Save to Drive
output_path = "/content/drive/MyDrive/generated_image.png"
image.save(output_path)
print(f"✅ Image saved to: {output_path}")
```
