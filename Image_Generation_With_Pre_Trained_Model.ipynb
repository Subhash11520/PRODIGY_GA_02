```python
from diffusers import StableDiffusionPipeline
import torch

# Load pre-trained Stable Diffusion model
pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=torch.float16,  # Faster if your GPU supports it
).to("cuda")
```

```python
prompt = "A futuristic cityscape at sunset, ultra-realistic, 4K"
image = pipe(prompt).images[0]

# Show image in notebook
image.show()

# Save image locally
image.save("generated_image.png")
```

```python
pipe = StableDiffusionPipeline.from_pretrained(
    "stabilityai/stable-diffusion-2-1",
    revision="fp16",
    torch_dtype=torch.float16,
).to("cuda")

# Then use the same prompt call
```

```python
prompt = "A dreamy alien jungle, cinematic lighting"
results = pipe(prompt, num_inference_steps=50, num_images_per_prompt=3)

for i, img in enumerate(results.images):
    img.save(f"image_{i}.png")
```

```python
from PIL import ImageDraw, ImageFont

def add_text_overlay(image, text):
    draw = ImageDraw.Draw(image)
    font = ImageFont.load_default()  # Uses default font

    text_position = (10, 10)
    text_color = (255, 255, 255)
    shadow_color = (0, 0, 0)

    draw.text((text_position[0]+1, text_position[1]+1), text, font=font, fill=shadow_color)
    draw.text(text_position, text, font=font, fill=text_color)

    return image

# Apply it
image_with_text = add_text_overlay(image, prompt)
image_with_text.show()
image_with_text.save("generated_with_text.png")
```

```python
import random

subjects = ["a medieval castle", "a cyberpunk street", "a floating island"]
modifiers = ["at sunset", "in the rain", "under moonlight"]
styles = ["in the style of Studio Ghibli", "ultra-realistic", "low poly 3D"]

def generate_prompt():
    return f"{random.choice(subjects)} {random.choice(modifiers)}, {random.choice(styles)}"

# Example:
prompt = generate_prompt()
image = pipe(prompt).images[0]
```

```python
import gradio as gr

def generate_image(prompt):
    image = pipe(prompt).images[0]
    return image

gr.Interface(
    fn=generate_image,
    inputs="text",
    outputs="image",
    title="Stable Diffusion Generator",
    description="Enter a prompt to generate an image"
).launch()
```
